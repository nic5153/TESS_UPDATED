```python
import csv

def create_csv_with_links_and_sector(input_csv_path, output_csv_path, target_category, base_link):
    # Read data from the input CSV file
    with open(input_csv_path, 'r') as input_file:
        reader = csv.DictReader(input_file)
        data = list(reader)

    print("Data from input CSV:")
    for row in data:
        print(row)

    # Extract data for rows containing the target category
    snia_data = [{'Sector': row['sector'], 'Type': row['type'], 'Name': row['name'], 'Link': f"{base_link}sector{row['sector']}/lc_{row['name']}_cleaned.png"} for row in data if target_category in row['type']]

    print("\nFiltered SNIA Data:")
    for entry in snia_data:
        print(entry)

    # Check if there are any matching entries
    if snia_data:
        # Write data to the output CSV
        with open(output_csv_path, 'w', newline='') as output_file:
            fieldnames = ['Sector', 'Type', 'Name', 'Link']
            writer = csv.DictWriter(output_file, fieldnames=fieldnames)

            # Write header
            writer.writeheader()

            # Write data
            writer.writerows(snia_data)

        print(f'Data for category "{target_category}" has been exported to {output_csv_path}')
    else:
        print(f'No data found for category "{target_category}" in the input CSV file.')

# Example usage:
input_csv_path = 'count_transients_catagorized.csv'
output_csv_path = 'sorted_snia_data.csv'
target_category_to_filter = 'SNIa'
base_link = 'https://tess.mit.edu/public/tesstransients/images/'

create_csv_with_links_and_sector(input_csv_path, output_csv_path, target_category_to_filter, base_link)
```

```python
#this code is for combining all the individual classified csv files into one csv

import csv
import os

def combine_csv_files(input_files, output_csv_path):
    # List to store data from all CSV files
    combined_data = []

    # Iterate through each input CSV file
    for input_file in input_files:
        # Read data from the current input CSV file
        with open(input_file, 'r') as file:
            reader = csv.DictReader(file)
            data = list(reader)
            combined_data.extend(data)

    # Write combined data to the output CSV file
    with open(output_csv_path, 'w', newline='') as output_file:
        fieldnames = combined_data[0].keys() if combined_data else []
        writer = csv.DictWriter(output_file, fieldnames=fieldnames)

        # Write header
        writer.writeheader()

        # Write combined data
        writer.writerows(combined_data)

    print(f'Data from {len(input_files)} CSV files has been combined and exported to {output_csv_path}')

input_files = ['/work/sorted_csv/sorted_snii_data.csv', '/work/sorted_csv/sorted_cv_data.csv', '/work/sorted_csv/sorted_agn_data.csv', '/work/sorted_csv/sorted_snia_data.csv', '/work/sorted_csv/sorted_snib_data.csv']
output_csv_path = 'classified_data.csv'

combine_csv_files(input_files, output_csv_path)
```

```python
#This code creates a notebook for each sector to plot the interesting light curves in that specific sector
import os
from nbformat import write, v4 as nbf

# Define the existing directory where you want to create the files
directory = r"C:\Users\nic51\TESS RESEARCH\Interesting_curves"

# Define the number of files you want to create
num_files = 66

# Loop through and create the files
for i in range(1, num_files + 1):
    # Define the filename
    filename = f"sector_{i}.ipynb"
    # Create a new Jupyter notebook object
    nb = nbf.new_notebook()
    # Write the notebook to a file
    with open(os.path.join(directory, filename), "w") as f:
        write(nb, f)
```

```python
import os

# Directory path for interesting curves
interesting_curves_directory = r'C:\Users\nic51\Code_Repos\Tess\Data\Interesting_Curves'

# List of subdirectories to create
subdirectories = ['CV', 'SN', 'Eclipses', 'Pulsators', 'Star Stuff', 'Super Outbursts','Other_Unknown']

# Iterate through each sector folder
for sector_folder in os.listdir(interesting_curves_directory):
    sector_folder_path = os.path.join(interesting_curves_directory, sector_folder)
    if os.path.isdir(sector_folder_path):
        # Create subdirectories within the sector folder
        for subdirectory in subdirectories:
            subdirectory_path = os.path.join(sector_folder_path, subdirectory)
            if not os.path.exists(subdirectory_path):
                os.makedirs(subdirectory_path)
                print(f"Created subdirectory '{subdirectory}' in '{sector_folder}' folder.")
            else:
                print(f"Subdirectory '{subdirectory}' already exists in '{sector_folder}' folder.")
    else:
        print(f"Skipping '{sector_folder}' as it's not a directory.")

print("Process completed successfully.")
```

```python
#to search for a light curve and display
import os

def search_light_curve(png_folder, curve_id):
    """
    Search for a specific light curve PNG file based on its ID.
    
    Args:
    - png_folder (str): Path to the directory containing PNG files.
    - curve_id (str): ID of the light curve to search for.
    
    Returns:
    - str: Path to the PNG file if found, None otherwise.
    """
    
    for filename in os.listdir(png_folder):
        if filename.endswith('.png'):
            
            file_id = filename.split('.')[0]
            
            if file_id == curve_id:
                
                return os.path.join(png_folder, filename)
    # If curve_id not found, return None
    return None

# Example usage:
png_folder = r'C:\Users\nic51\Code_Repos\Tess\Data\Extracted_Tarballs\PNG_FILES'
curve_id = '2020nhc'
result = search_light_curve(png_folder, curve_id)
if result:
    print(f"Light curve PNG file found: {result}")
else:
    print(f"No light curve PNG file found for ID: {curve_id}")
```
