```python

import numpy as np
from astropy.timeseries import LombScargle
import matplotlib.pyplot as plt


time = np.linspace(0, 10, 100)
amplitude = 2
frequency = 2

data = amplitude * np.sin(2 * np.pi * frequency * time)


ls = LombScargle(time, data)


frequency = np.linspace(0.1, 5, 1000)

# Calculate and plot the Lomb-Scargle periodogram
power = ls.power(frequency)
plt.plot(frequency, power)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Lomb-Scargle Power")
plt.title("Lomb-Scargle Periodogram for Sine Wave")
plt.xlim(0, 5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import os
from scipy.signal import find_peaks
from astropy.timeseries import LombScargle

data_file = r"C:\Users\nic51\Code_Repos\Tess\Data\Interesting_Curves\sector07\Super Outbursts\2019agu.txt"

data = np.loadtxt(data_file)
time = data[:, 1]
magnitude = data[:, 4]
ls = LombScargle(time, magnitude)

frequency = np.linspace(.1, 10, 1000)
power = ls.power(frequency)


plt.plot(frequency,power)
plt.ylim(0, max(power) * 1.1)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Lomb-Scargle Power")

peaks, _ = find_peaks(power, height=0)

print("Significant peaks:")
for peak_index in peaks:
    peak_frequency = frequency[peak_index]
    peak_power = power[peak_index]
    print(f"Frequency: {peak_frequency:.2f} Hz, Power: {peak_power:.2f}")
    
plt.plot(frequency[peaks], power[peaks], 'ro', markersize=5, label='Peaks')
plt.legend()

specific_frequency = 6.47
span_days = 27
period = 1 / specific_frequency
cycles_per_day = specific_frequency * span_days
print("Frequency in Hertz:", specific_frequency)
print("Span of data in days:", span_days)
print("Cycles per day:", cycles_per_day)


plt.show()
```
